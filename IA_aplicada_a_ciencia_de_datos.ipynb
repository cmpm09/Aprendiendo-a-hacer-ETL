## Como crear un buen prompt con chat gpt

Una de las habilidades más importantes para la interacción con IAs generativas como ChatGPT es la capacidad de crear buenos prompts, guiando la herramienta de acuerdo con tu conocimiento sobre el negocio. La creación de un prompt eficaz en ChatGPT es fundamental para obtener resultados precisos y relevantes para la resolución de problemas.

En el contexto de un proyecto de análisis de datos, como el que estamos trabajando aquí, un prompt bien elaborado puede ser la clave para obtener insights valiosos. A continuación, se presentan algunas buenas prácticas y ejemplos, teniendo en cuenta un proyecto simple de Ciencia de Datos:

**Sé específico y directo:**

Explica directamente lo que deseas que ChatGPT ejecute. Esto ayuda a la IA a entender mejor tu necesidad, proporcionando respuestas más precisas.

Ejemplo: “Realiza un análisis exploratorio de datos en la base ‘zoop_ventas.csv' usando las bibliotecas de manipulación de datos de Python. Presenta un resumen estadístico de las ventas por región”.

**Proporciona contexto cuando sea necesario:**

Informa a ChatGPT del contexto del proyecto, incluyendo datos, objetivos y restricciones. Esto puede mejorar significativamente la calidad de las respuestas. Sin embargo, evita información innecesaria que pueda confundir a la IA.

Ejemplo: "Estoy trabajando en un proyecto de análisis de ventas de una tienda minorista. La base de datos contiene información de ventas diarias por categoría de producto y región en 2023. Ayúdame a calcular la suma de las ventas de 'Electrónicos' por trimestre".

**Especifica el formato de la respuesta:**

Indica el formato deseado para la respuesta, como gráficos, tablas o explicaciones detalladas.

Ejemplo: "Crea un gráfico de líneas que muestre la evolución de las ventas a lo largo del último año, segmentado por categoría de productos".

**Indica el nivel de detalle deseado:**

Deja claro el nivel de detalle o profundidad de la explicación que estás buscando. Esto ayuda a la IA a ajustar la respuesta a tu necesidad de comprensión.

Ejemplo: "¿Podrías proporcionarme un ejemplo comentado de cómo usar la función groupby() de pandas para agrupar datos por categoría y calcular la suma total de ventas para cada grupo?"

**Usa términos técnicos correctos:**

Pon atención a la terminología específica relacionada con tu área de interés. Esto asegura que la IA comprenda el dominio de tu pregunta y responda de forma adecuada.

Ejemplo: “¿Cómo implemento un modelo de regresión lineal en Python usando scikit-learn para predecir ventas futuras con base en gastos de publicidad?”

**Divide tareas complejas en etapas simples:**

Si el proyecto involucra varias etapas, divídelo en partes más pequeñas e instruye a ChatGPT paso a paso. Esto facilita que la IA te ayude de manera eficaz.

Ejemplo:

"¿Cómo puedo manejar valores faltantes en mi conjunto de datos usando pandas?"

"¿Cuál es la mejor manera de visualizar la distribución de edades en mi conjunto de datos usando matplotlib?"

“¿Podrías ajustar el visual de la distribución de edades colocando las etiquetas y labels con tamaño 12 y color #CCCCCC?”

**Mantente abierto a interacciones:**

Esté siempre preparado(a) para realizar ajustes en tus prompts basados en las respuestas recibidas. La interacción puede ayudar a refinar las respuestas y obtener exactamente lo que necesitas.

Ejemplo: “Esta respuesta puede mejorar. ¿Qué tal si redondeamos los valores a dos decimales y mostramos en el gráfico solo los valores mayores a 1 millón (1e6)?”

Estos consejos son fundamentales en la construcción de prompts bien instruidos, como los que encontrarás a lo largo de este curso. Observa cómo se describen los pasos y acciones en los prompts del curso y cómo puedes aplicarlos en tu proyecto utilizando ChatGPT.

Asegúrate de comunicar tus intenciones de forma precisa, dividir tareas complejas en etapas manejables y proporcionar ejemplos cuando sea apropiado. Esto te ayudará a obtener respuestas más relevantes y útiles para tu proyecto, haciendo la interacción con el chat más eficaz.

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear una nueva columna con la facturación total por fila
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# Agrupar por categoría y sumar la facturación
facturacion_categoria = df.groupby('categoria')['facturacion'].sum().sort_values(ascending=False).reset_index()

# Crear gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(data=facturacion_categoria, y='categoria', x='facturacion', palette='coolwarm', orient= 'h')

# Añadir detalles al gráfico
plt.title('Facturación por Categoría de Producto')
plt.xlabel('Facturación Total (en pesos mexicanos)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
````

### **3a Pregunta:** ¿Cómo fueron las ventas a lo largo de los meses en 2023?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurar que la columna fecha esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear columna auxiliar con el año y mes (periodo mensual)
df['mes_periodo'] = df['fecha'].dt.to_period('M')

# Agrupar por mes y sumar la facturación
ventas_mensuales = df.groupby('mes_periodo')['facturacion'].sum().reset_index()

# Convertir el periodo a datetime para formateo
ventas_mensuales['mes_datetime'] = ventas_mensuales['mes_periodo'].dt.to_timestamp()

# Crear columna "mes" con el nombre traducido
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}
ventas_mensuales['mes'] = ventas_mensuales['mes_datetime'].dt.month_name().map(meses)

# Ordenar por fecha para que la línea tenga sentido cronológico
ventas_mensuales = ventas_mensuales.sort_values('mes_datetime')

# Crear gráfico de líneas
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o', linewidth=2.5, color='teal')

# Añadir detalles
plt.title('Ventas Totales Mensuales - Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación Total')
plt.grid(True)
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

Cuando tenemos una cantidad mayor de períodos en nuestros datos, vale la pena recurrir a gráficos de líneas, que son ideales para series temporales. Dado que nuestros datos son mensuales, utilizar un gráfico de columnas, por ejemplo, reduciría la percepción de la secuencia de los datos. Las líneas ofrecen una mayor sensación de continuidad.

Vamos a elegir ahora el gráfico de líneas (Ventas totales mensuales de Zoop) y aplicar una técnica más a él mediante el uso de la IA.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurar que la columna fecha esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear columna auxiliar con el año y mes (periodo mensual)
df['mes_periodo'] = df['fecha'].dt.to_period('M')

# Agrupar por mes y sumar la facturación
ventas_mensuales = df.groupby('mes_periodo')['facturacion'].sum().reset_index()

# Convertir el periodo a datetime para formateo
ventas_mensuales['mes_datetime'] = ventas_mensuales['mes_periodo'].dt.to_timestamp()

# Crear columna "mes" con el nombre traducido
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}
ventas_mensuales['mes'] = ventas_mensuales['mes_datetime'].dt.month_name().map(meses)

# Ordenar por fecha para que la línea tenga sentido cronológico
ventas_mensuales = ventas_mensuales.sort_values('mes_datetime')

# Crear gráfico de líneas
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o', linewidth=2.5, color='teal')

# Añadir detalles
plt.title('Ventas Totales Mensuales - Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación Total')
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Diccionario de traducción de meses
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

# Crear columna de nombre de mes y calcular ventas mensuales
df['mes'] = df['fecha'].dt.strftime('%B')
ventas_mensuales = df.groupby('mes')['facturacion'].sum().reindex(meses.keys())

# Crear la figura
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor(GRIS_1)
ax.set_facecolor(GRIS_1)

# Graficar línea
ax.plot(list(meses.values()), ventas_mensuales.values, marker='o', color=AZUL_1, linewidth=2)

# Título
ax.set_title("Ventas totales mensuales en 2023", fontsize=18, color=GRIS_5)

# Formato eje Y: MXN$ valor M
formatter = ticker.FuncFormatter(lambda x, _: f'MXN$ {x / 1e6:.2f} M')
ax.yaxis.set_major_formatter(formatter)
ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

# Eje X
ax.set_xticks(range(12))
ax.set_xticklabels(list(meses.values()), fontsize=12, color=AZUL_1)

# Suprimir etiquetas de los ejes
ax.set_xlabel('')
ax.set_ylabel('')

# Quitar bordes superior y derecho
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.patches import Circle

# Diccionario de traducción de meses
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

# Crear columna de nombre de mes y calcular ventas mensuales
df['mes'] = df['fecha'].dt.strftime('%B')
ventas_mensuales = df.groupby('mes')['facturacion'].sum().reindex(meses.keys())

# Crear la figura
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor(GRIS_1)
ax.set_facecolor(GRIS_1)

# Graficar línea de ventas
mes_labels = list(meses.values())
ventas = ventas_mensuales.values
ax.plot(mes_labels, ventas, marker='o', color=AZUL_1, linewidth=2)

# Título
ax.set_title("Ventas totales mensuales en 2023", fontsize=18, color=GRIS_5)

# Formato eje Y
formatter = ticker.FuncFormatter(lambda x, _: f'MXN$ {x / 1e6:.2f} M')
ax.yaxis.set_major_formatter(formatter)
ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

# Eje X
ax.set_xticks(range(12))
ax.set_xticklabels(mes_labels, fontsize=12, color=AZUL_1)

# Quitar bordes superiores y derechos
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Suprimir labels de ejes
ax.set_xlabel('')
ax.set_ylabel('')

# Coordenadas de los puntos que vamos a encerrar (5º, 8º, 11º y 12º => índices 4, 7, 10, 11)
puntos_destacados = [4, 7, 10, 11]

for i, idx in enumerate(puntos_destacados):
    x = idx
    y = ventas[idx]
    # Añadir círculo visible
    circulo = Circle((x, y), radius=0.4, fill=False, edgecolor=ROJO_1, linewidth=2, transform=ax.transData)
    ax.add_patch(circulo)
    # Enumeración a la izquierda del círculo
    ax.text(x - 0.5, y, str(i + 1), fontsize=12, color=ROJO_1, fontweight='bold', ha='right', va='center')

plt.tight_layout()
plt.show()

````

````python
def distribucion_calificaciones():
  import matplotlib.pyplot as plt
  import seaborn as sns

  # 1. Crear figura y ejes
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 2. Crear histograma
  sns.histplot(df['nota'], bins=11, kde=False, color=ROJO_1, ax=ax)

  # 3. Título del gráfico
  ax.set_title('Distribución de las evaluaciones de las compras en 2023', fontsize=18, color=GRIS_5)

  # 4. Etiquetas de los ejes
  ax.set_xlabel('Calificación (0-10)', fontsize=12, color=AZUL_1)
  ax.set_ylabel('Frecuencia', fontsize=12, color=AZUL_1)

  # 5. Personalizar ticks
  ax.set_xticks(range(11))  # De 0 a 10
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # 6. Quitar bordes del gráfico
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # 7. Agregar grid en el eje y (detrás del gráfico)
  ax.grid(axis='y', linestyle='--', linewidth=0.5)
  ax.set_axisbelow(True)

  # 8. Preparar datos para anotación
  promedio = df['nota'].mean().round(2)
  nota_mas_comun = df['nota'].value_counts().idxmax()
  frecuencia_mas_comun = df['nota'].value_counts().max()

  # 9. Agregar caja de anotación
  texto_anotacion = (
      f"Promedio de calificaciones = {promedio}\n"
      f"Calificación más común = {nota_mas_comun}\n"
      f"Frecuencia más común = {frecuencia_mas_comun}"
  )

  ax.annotate(
      texto_anotacion,
      xy=(8.7, 2250),
      xytext=(2, 1800),
      fontsize=12,
      color=AZUL_1,
      arrowprops=dict(arrowstyle="->", color=AZUL_1),
      bbox=dict(boxstyle="round,pad=0.4", facecolor=BLANCO, edgecolor=GRIS_3)
  )

  plt.tight_layout()
  return plt
````

````python
def box_plot_edad_sexo():
  import matplotlib.pyplot as plt
  import seaborn as sns

  # 1. Crear figura y ejes
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 2. Crear el boxplot
  sns.boxplot(
      x='sexo_biologico',
      y='edad',
      data=df,
      palette=[AZUL_1, AQUA_1],
      ax=ax
  )

  # 3. Título
  ax.set_title('Distribución de la Edad por Sexo Biológico de los clientes', fontsize=18, color=GRIS_5)

  # 4. Suprimir etiqueta del eje x
  ax.set_xlabel(None)

  # 5. Etiquetas y ticks
  ax.set_ylabel('Edad', fontsize=12, color=AZUL_1)
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # 6. Quitar bordes del gráfico
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # 7. Grid detrás del gráfico
  ax.grid(axis='y', linestyle='--', linewidth=0.5)
  ax.set_axisbelow(True)

  # 8. Calcular estadísticas por grupo
  agrupado = df.groupby('sexo_biologico')['edad']
  estadisticas = agrupado.describe(percentiles=[0.25, 0.5, 0.75])

  # 9. Anotación para el grupo índice 1 (por ejemplo, 'Masculino')
  texto1 = (
      f"Mínimo = {int(estadisticas['min'][1])}\n"
      f"1º quartil = {int(agrupado.quantile(0.25)[1])}\n"
      f"2º quartil = {int(estadisticas['50%'][1])}\n"
      f"3º quartil = {int(agrupado.quantile(0.75)[1])}\n"
      f"Máximo = {int(estadisticas['max'][1])}"
  )

  ax.annotate(
      texto1,
      xy=(-0.1, 45), xytext=(-0.4, 50),
      fontsize=12,
      color=AZUL_1,
      arrowprops=dict(arrowstyle="->", color=AZUL_1),
      bbox=dict(boxstyle="round,pad=0.4", facecolor=BLANCO, edgecolor=GRIS_3)
  )

  # 10. Anotación para el grupo índice 0 (por ejemplo, 'Femenino')
  texto2 = (
      f"Mínimo = {int(estadisticas['min'][0])}\n"
      f"1º quartil = {int(agrupado.quantile(0.25)[0])}\n"
      f"2º quartil = {int(estadisticas['50%'][0])}\n"
      f"3º quartil = {int(agrupado.quantile(0.75)[0])}\n"
      f"Máximo = {int(estadisticas['max'][0])}"
  )

  ax.annotate(
      texto2,
      xy=(0.9, 45), xytext=(0.6, 50),
      fontsize=12,
      color=AZUL_1,
      arrowprops=dict(arrowstyle="->", color=AZUL_1),
      bbox=dict(boxstyle="round,pad=0.4", facecolor=BLANCO, edgecolor=GRIS_3)
  )

  plt.tight_layout()
  return plt
````
